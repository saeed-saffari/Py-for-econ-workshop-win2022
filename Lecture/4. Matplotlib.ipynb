```python
x = np.linspace(-10,10,50)
#print(x)
y = x ** 2
#print(y)

plt.figure(figsize=(12,7))ترم 
plt.plot(x,y,
         c='navy', ls = '-', lw = 3, marker = 'o', markersize = 12, markeredgecolor = 'g',
        markerfacecolor = 'm', label = '$y=x^2$')

plt.legend(fontsize = 12,loc = 'upper center')

plt.grid(c = 'k', ls = '--', alpha = 0.5)

plt.title('Function $Y = x^2$\nin range (-10,10)', fontsize = 15, fontweight = 'bold',
          fontname = 'Times New Roman', c = 'darkblue')

plt.xlabel('X range', fontsize = 12, fontweight = 'bold',fontname = 'Times New Roman')
plt.ylabel('Y range', fontsize = 12, fontweight = 'bold',fontname = 'Times New Roman')

#plt.text(-10,-20,'First text\nin figure!', fontsize = 12,fontname = 'Times New Roman')

#plt.xlim(left = -7.5,right = 0)
#plt.ylim(bottom = 20,top = 80)

plt.axhline(40, c= 'k', ls= '--')
#plt.axhline(75, c= 'k', ls= '--')
plt.axvline(0, c= 'g', ls= '--')

#plt.savefig('test_1.jpg',dpi = 700)

plt.show()
```

```python
x = np.linspace(-10,10,50)
y = x ** 2

plt.figure(figsize=(12,7))
plt.plot(x, 10*y, c = 'r', ls = '-', lw = 3, label = '$y = x^2$')
plt.plot(x, x ** 3,c = 'g', ls = '-.', lw = 3, label = '$y = x^3$')
plt.plot(x, 150 * np.sin(x),c = 'b', ls = '--', lw = 3, label = '$y = sin(x)$')

plt.legend(fontsize = 12, loc = 'best')
plt.grid()

plt.show()
```

```python
plt.figure(figsize = (15,8))

plt.subplot(2,3,1)
plt.plot(x,x**2)

plt.subplot(2,3,2)
plt.plot(x,x**3)
plt.title('$y = x^3$')

plt.subplot(2,3,3)
plt.plot(np.sin(x))

plt.subplot(2,3,5)
plt.plot(x,np.tan(x))
plt.grid()

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (12,7))

ax.plot(x, x**2, label = 'ONE')
ax.plot(x, 50*np.sin(x), label = 'TWO')

ax.grid()
ax.legend()

ax.set_title('Compare')
ax.set_xlabel('X range')
ax.set_ylabel('Y range')

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2,1, figsize = (12,7), sharex=True)

ax1.plot(x, x**2, label = 'ONE', c = 'r')
ax2.plot(x, 50*np.sin(x), label = 'TWO', c = 'b')

ax1.grid()
ax2.grid()

ax1.legend()
ax2.legend()

ax2.set_xlabel('X range')
#ax1.set_xlabel('X range')
ax1.set_ylabel('Y range')
ax2.set_ylabel('Y range')

ax1.set_title('Compare')

plt.show()
```

$$ y = 20 + x $$
$$ y = 100 - 2x $$


```python
# example

x = np.linspace(0,100,100)
y_1 = 20 + x
y_2 = 100 - 2 * x

plt.figure(figsize=(12,7))

plt.plot(x,y_1, c = 'r', lw = 2, label = 'Supply')
plt.plot(x,y_2, c = 'b', lw = 2, label = 'Demand')

# demand after raxation
plt.plot(x,y_2 - 20, c = 'g',ls = '-.', lw = 2, label = 'Demand after taxation')

plt.ylim(bottom = 0)
plt.xlim(left = 0)

plt.legend()
plt.grid()

plt.axhline(80/3 +20, c = 'k', ls = '--', lw = 1.5)
plt.axvline(80/3, c = 'k', ls = '--', lw = 1.5)

plt.axhline(60/3 +20, c = 'm', ls = '--', lw = 1.5)
plt.axvline(60/3, c = 'm', ls = '--', lw = 1.5)

plt.text(80/3+1, 80/3+30, 'Equa= 12')

plt.xlabel('Quantity')
plt.ylabel('Price')
plt.title('Demand and Suply plot')

plt.fill_between(x,y_2,y_1, where= (y_2 > y_1) & (x>=20), color = 'darkorange', alpha = 0.7)
plt.show()
```

## Bar

```python
plt.figure(figsize=(12,7))

plt.bar([1,3,5,7],[5,8,2,6], label = 'ONE')
plt.bar([2,4,6,8],[1,4,8,3], label = 'TWO')

plt.grid(axis = 'y', c = 'k')
plt.legend()
plt.show()
```

## Scatter

```python
x = np.linspace(0,10,100)
e = np.random.rand(100) * 100
y = x + e

scale = np.random.rand(100) * 1000
color = np.random.rand(100)


plt.figure(figsize=(12,7))
plt.scatter(x,y, s = scale, c = color)

plt.colorbar().set_label('\nColor bar', fontsize = 15)

plt.grid()
plt.show()
```

```python
z = np.arange(100)
d = np.arange(100) * 2
m = np.arange(100)
a = np.random.rand(100)
b = z + a * 20

plt.figure(figsize=(12,7))
plt.scatter(z,b, c= m, s = d)
plt.colorbar()
plt.grid()
plt.show()
```

```python
data = np.random.normal(0,1,100000)

x = np.linspace(-4,4,100)
y = 1/np.sqrt(2 * np.pi) * np.exp(-x**2/2)

plt.figure(figsize=(12,7))
plt.plot(x,y,lw =3,c='k')
plt.hist(data,70, density=1)

plt.show()
```
